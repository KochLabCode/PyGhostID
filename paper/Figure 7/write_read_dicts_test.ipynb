```python
import pickle
import numpy as np
myDict = {'name': 'Alice', 'age': 30, 'city': 'New York', 'height': 5.5, "scores": np.array([85, 90, 95])}
test = {'key1': 'value1', 'key2': 42}
list = [[myDict, test], [test, myDict]]
with open('test.pkl', 'wb') as f:
    pickle.dump(list, f) 

with open('test.pkl', 'rb') as f:
    list2 = pickle.load(f)

print(list2)
```
